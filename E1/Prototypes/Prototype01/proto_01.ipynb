```python
# Python 3.8.12

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from ultralytics import YOLO
```

```python

model = YOLO('yolov8n.pt')
```

```python
bob = model.predict(
   source='https://media.roboflow.com/notebooks/examples/dog.jpeg',
   conf=0.25
)
```
